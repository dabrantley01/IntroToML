## Clean up the data


As a first pass we don't optimize our data. We take the raw data, sans (possibly relevant) columns, and build classification models with the limited feature space.

For this limited feature space, we consider only Age, Sex, Fare and Pclass as predictors for Survived. 

```python
dfnew = df[["Age","Sex","Pclass","Fare","Survived"]]
dfnew.head()
```

Drop rows with NaNs.

```python
dfnew.isnull().sum()
```

```python
dfnew = dfnew[dfnew["Age"].isnull() == False]
```

At the very least we should hot-fix our categorical columns.

```python
y = dfnew['Survived']

raw_features = dfnew.drop('Survived',axis = 1)

# 1-hot encoding of categorical columns "Pclass" and "Sex".
X = pd.get_dummies(raw_features,columns=["Sex","Pclass"])

X.head()
```

## Classification Model Testing and Evaluation.


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, classification_report
from matplotlib.colors import ListedColormap
```

```python
# Scale and split data.
X = StandardScaler().fit_transform(X)
X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=.2, random_state=42)
```

## Decision Tree Classifier

A decision tree classification operates by building split-levels between features and assigning probabilities to each split.

```python
# Fit the model.
clf = DecisionTreeClassifier(max_depth=10)
clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
name = 'Decision Tree'

fig, ax = plt.subplots(figsize=(10,7))

x1_min, x1_max = X[:, 0].min(), X[:, 0].max()
x2_min, x2_max = X[:, 1].min(), X[:, 1].max()
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, h),
                     np.arange(x2_min, x2_max, h))

# just plot the dataset first
cm = plt.cm.RdBu
cm_bright = ListedColormap(['#FF0000', '#0000FF'])

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
if hasattr(clf, "decision_function"):
    mesh_test_pts = np.c_[xx1.ravel(), 
                          xx2.ravel(),
                          np.repeat(np.median(X[:,2]),len(xx1.ravel())),
                          np.repeat(np.median(X[:,3]),len(xx1.ravel())),
                          np.repeat(np.median(X[:,4]),len(xx1.ravel())),
                          np.repeat(np.median(X[:,5]),len(xx1.ravel())),
                          np.repeat(np.median(X[:,6]),len(xx1.ravel()))
                         ]
    Z = clf.decision_function(mesh_test_pts)
else:
    mesh_test_pts = np.c_[xx1.ravel(), 
                          xx2.ravel(),
                          np.repeat(np.median(X[:,2]),len(xx1.ravel())),
                          np.repeat(np.median(X[:,3]),len(xx1.ravel())),
                          np.repeat(np.median(X[:,4]),len(xx1.ravel())),
                          np.repeat(np.median(X[:,5]),len(xx1.ravel())),
                          np.repeat(np.median(X[:,6]),len(xx1.ravel()))
                         ]

    Z = clf.predict_proba(mesh_test_pts)[:, 1]

# Put the result into a color plot
Z = Z.reshape(xx.shape)
ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)

# Plot the training points
ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
           edgecolors='k')
# Plot the testing points
ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
           edgecolors='k', alpha=0.6)

ax.set_xlim(xx.min(), xx.max())
ax.set_ylim(yy.min(), yy.max())
ax.set_xticks(())
ax.set_yticks(())
ax.set_title(name)
ax.text(xx.max() - .3, yy.max() - 1.0, ('%.2f' % score).lstrip('0'),
        size=15, horizontalalignment='right')
```

```python
# Define our classifier objects.
names = ["Nearest Neighbors", "Linear SVM", "RBF SVM", "Gaussian Process",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]

classifiers = [
    KNeighborsClassifier(5),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=10),
    RandomForestClassifier(max_depth=10, n_estimators=10, max_features="auto"),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()]
```

```python
# Calculate then plot data and training results.
fig,axarr = plt.subplots(ncols=1,nrows=len(classifiers) + 1,figsize=(9, 35))

h = .02  # step size in the mesh

x1_min, x1_max = X[:, 0].min(), X[:, 0].max()
x2_min, x2_max = X[:, 1].min(), X[:, 1].max()
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, h),
                     np.arange(x2_min, x2_max, h))

# just plot the dataset first
cm = plt.cm.RdBu
cm_bright = ListedColormap(['#FF0000', '#0000FF'])

ax.set_title("Input data")

# Plot the training points.
axarr[0].scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
           edgecolors='k')

# Plot the testing points.
axarr[0].scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6,
           edgecolors='k')
axarr[0].set_xlim(xx.min(), xx.max())
axarr[0].set_ylim(yy.min(), yy.max())
axarr[0].set_xticks(())
axarr[0].set_yticks(())

i = 1
for clf,name in zip(classifiers,names):
    ax = axarr[i]

    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    if hasattr(clf, "decision_function"):
        mesh_test_pts = np.c_[xx1.ravel(), 
                              xx2.ravel(),
                              np.repeat(np.median(X[:,2]),len(xx1.ravel())),
                              np.repeat(np.median(X[:,3]),len(xx1.ravel())),
                              np.repeat(np.median(X[:,4]),len(xx1.ravel())),
                              np.repeat(np.median(X[:,5]),len(xx1.ravel())),
                              np.repeat(np.median(X[:,6]),len(xx1.ravel()))
                             ]
        Z = clf.decision_function(mesh_test_pts)
    else:
        mesh_test_pts = np.c_[xx1.ravel(), 
                              xx2.ravel(),
                              np.repeat(np.median(X[:,2]),len(xx1.ravel())),
                              np.repeat(np.median(X[:,3]),len(xx1.ravel())),
                              np.repeat(np.median(X[:,4]),len(xx1.ravel())),
                              np.repeat(np.median(X[:,5]),len(xx1.ravel())),
                              np.repeat(np.median(X[:,6]),len(xx1.ravel()))
                             ]

        Z = clf.predict_proba(mesh_test_pts)[:, 1]

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)

    # Plot the training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
               edgecolors='k')
    # Plot the testing points
    ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
               edgecolors='k', alpha=0.6)

    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(name)
    ax.text(xx.max() - .3, yy.max() - 1.0, ('%.2f' % score).lstrip('0'),
            size=15, horizontalalignment='right')
    i += 1
```
